```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./Car_sales.csv')
```

```python
df
```

```python
print("Средняя цена автомобиля: ",(df['Price_in_thousands'].mean()*1000).round(2),'$')
print("Средний объем двигателя: ",df['Engine_size'].mean().round(2),'L')
```

```python
print("3 самых популярных авто")
df.sort_values('Sales_in_thousands',ascending=False).head(3)
```

```python
print("Авто с самым маленьким расходом топлива")
df.sort_values('Fuel_efficiency',ascending=False).head(3)
```

```python
print("Средняя цена на авто марки BMW:",df[df['Manufacturer']=="BMW"]['Price_in_thousands'].mean().round()*1000,'$')
```

```python
print("Авто с самым мощным двигателем")
df[df['Horsepower']==df['Horsepower'].max()]
```

```python
print("Кол-во проданных авто по производителям")
for_round=df.groupby(['Manufacturer'])['Sales_in_thousands'].sum()
for_round
```

```python
fig1, ax1 = plt.subplots()
 
wedges, texts, autotexts = ax1.pie(for_round, labels=for_round.index, autopct='%1.2f%%')
ax1.axis('equal')
ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
for_plot=df[df['Manufacturer']=="Ford"]
for_plot
```

```python
print("Продажи авто марки Ford")
fig1, ax1 = plt.subplots()
wedges, texts, autotexts = ax1.pie(for_plot['Sales_in_thousands'], labels=for_plot['Model'], autopct='%1.2f%%')
ax1.axis('equal')
ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.show()
```
